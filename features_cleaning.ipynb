```python
from datascience import *
import numpy as np
%matplotlib inline
import pandas as pd
```

```python
features = pd.read_csv("features.csv")
features
```

```python
#sets the first column as track id
features.iloc[1, 0] = 'track_id'
#sets the column names as the first row
features.columns = features.iloc[1]
#drop the first 3 rows
features = features.drop(features.index[0:3])
#only select track id and feature columns
features = features.iloc[:, [0,1,2,3,4,5,6,7,8]]
#features.drop(features.columns[9], axis = 1, inplace = True)
features
```

```python
features.dtypes
```

```python
#convert track id to integers
def make_int(x):
    if type(x) ==str:
        return int(x)
    else:
        return x
features['track_id'] = features['track_id'].apply(make_int).apply(int)
```

```python
#convert audio features to integers
def make_int_from_float(x):
    if type(x) ==str:
        return float(x)
    else:
        return x


cols = make_array('acousticness', 'danceability', 'energy',
       'instrumentalness', 'liveness', 'speechiness', 'tempo', 'valence')
for col in cols: 
    features[col] = features[col].apply(make_int_from_float).apply(float)
```
